```python
import tensorflow as tf
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
from collections import Counter
from scipy.stats import chi2_contingency
```

```python
allergies = pd.read_csv('allergies.csv')
patients = pd.read_csv('patients.csv')
```

```python
label_allergies = allergies['DESCRIPTION'].unique()
lst_allergies = allergies.DESCRIPTION
dct_allergies = Counter(lst_allergies)
d_allergies = dct_allergies.values()
d_nonallergies = len(patients) - len(allergies['PATIENT'].unique())
```

```python
plt.pie([d_nonallergies, len(allergies['PATIENT'].unique())], labels = ['Non-Allergic', 'Allergic'])
plt.title('Percentage of Patients with Allergies')
```

```python
dct_allergic = Counter(allergies.PATIENT)
d_allergic = dct_allergic.values()
plt.bar(allergies['PATIENT'].unique(), d_allergic)
plt.title('Number of Allergies')
plt.xlabel('Patient Code')
plt.ylabel('# of Allergies')
plt.show()
```

```python
plt.pie(d_allergies, labels = label_allergies )
plt.title('Percentage of Reported Allergies')
```

```python
ad_allergies = len(Counter(allergies.STOP))
took_allergies = len(Counter(allergies.START))
nad_allergies = took_allergies - ad_allergies
plt.pie([ad_allergies, nad_allergies], labels = ['Finished Prescription', 'Started but did not Finished' ])
plt.title('Percentage of Drug Adherence')
```
